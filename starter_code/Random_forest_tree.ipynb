```python
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter('ignore', FutureWarning)
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
df.columns
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
# selected_features = df[['names', 'of', 'selected', 'features', 'here']]
data = df 
selected_features = data.drop('koi_disposition',axis = 1)
selected_features.head()
X = selected_features
y = df['koi_disposition']
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler

X_scaler1 = MinMaxScaler().fit(X_train)

X_train_scaled1 = X_scaler1.transform(X_train)
X_test_scaled1 = X_scaler1.transform(X_test)
```

```python
#Building Model
from sklearn.ensemble import RandomForestClassifier
# Create a random forest classifier
model1 = RandomForestClassifier(n_estimators=200)

# Fit (Train) our model to the data

model1 = model1.fit(X_train_scaled1, y_train)

model1_training_score = (model1.score(X_train_scaled1, y_train))*100
base_accuracy = round((model1.score(X_test_scaled1, y_test))*100,3)

print(f"Training Score: {model1_training_score} %")
print(f"Testing Score: {base_accuracy} %")
```

```python
# Random Forests in sklearn will automatically calculate feature importance
importances = model1.feature_importances_

feature_names = X.columns.to_list()
# We can sort the features by their importance
all_features = sorted(zip(importances, feature_names), reverse=True)
features_df = pd.DataFrame(all_features,columns =['Ranks','Features'])

useful_features = features_df[features_df['Ranks'] > 0.01]['Features']
useful_features = useful_features.to_list()
display(all_features)
useful_features
```

```python
X_train_useful = X_train[useful_features]
X_test_useful = X_test[useful_features]
```

```python
#Scaling
X_scaler2 = MinMaxScaler().fit(X_train_useful)

X_train_scaled2= X_scaler2.transform(X_train_useful)
X_test_scaled2 = X_scaler2.transform(X_test_useful)
```

```python
model2 =RandomForestClassifier(n_estimators=200)

# Fit (Train) our model to the data
model2.fit(X_train_scaled2, y_train)
```

```python
useful_features_training_score = model2.score(X_train_scaled2, y_train)*100
useful_features_accuracy = round(model2.score(X_test_scaled2, y_test)*100,3)
```
